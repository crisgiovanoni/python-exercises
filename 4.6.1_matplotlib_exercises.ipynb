```python
import matplotlib.pyplot as plt
import math


x = range(-10,10)
y = [n**2 - n + 2 for n in x]

# fig = plt.figure()
# fig.add_subplot(221)

plt.plot(x,y)
plt.text(-1.5,10,'$y=x^2-x+2$',fontsize=10, color='blue')

plt.show()
```

```python
plt.figure(figsize=(5, 6))
plt.suptitle("Aggregation of Charts")

x = range(1,5)

#Subplot_1
y = [n**3 for n in x]

plt.subplot(221)
plt.plot(x,y)

#Subplot_2
y = [math.sqrt(n) for n in x]

plt.subplot(222)
plt.plot(x,y)

#Subplot_3
y = [math.tan(n) for n in x]

plt.subplot(223)
plt.plot(x,y)

#Subplot_4
y = [2**n for n in x]

plt.subplot(224)
plt.plot(x,y)

plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.suptitle("Aggregation of Charts")

x = range(1,5)

y1 = [n**3 for n in x]
y2 = [math.sqrt(n) for n in x]
y3 = [math.tan(n) for n in x]
y4 = [2**n for n in x]

plt.plot(x, y1, marker='o', markerfacecolor='gray', markersize=5, color='skyblue', linewidth=2, label="$y=x$")
plt.plot(x, y2, marker='o', markerfacecolor='gray', markersize=5, color='olive', linewidth=2, linestyle="dotted",label="$y=x^3$")
plt.plot(x, y3, marker='o', markerfacecolor='gray', markersize=5, color='orange', linewidth=2, linestyle="solid", label="$y=tan(x)$")
plt.plot(x, y4, marker='o', markerfacecolor='gray', markersize=5, color='pink', linewidth=2, linestyle="dashed", label="$y=2^x$")

plt.legend()
plt.show()
```
